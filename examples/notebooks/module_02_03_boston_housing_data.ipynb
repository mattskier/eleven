### A Bird's Eye View

One of my first goto visualizations is the histogram. This will give us a quick visual check to see if there are outliers.

```python
# Check out the distributions of our features
# Note our y-axes across the subplots. Which is more informative,
# sharing that y-axis information across subplots or not?
fig, axs = plt.subplots(ncols=7, nrows=2, figsize=(20, 10))
axs = axs.flatten()

for idx, feature in enumerate(X.T):
    axs[idx].hist(feature)
    axs[idx].set_title(f'{feature_names[idx]}')

# Since we have 2 rows each with 3 columns,
# and only 13 features, then we can visualize
# our target in the last `Axes` index.
axs[-1].hist(y, color='orange')
axs[-1].set_title(f'Target: MEDV')

plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

At first glance we can see that a few of these distributions are skewed. Though we haven't talked about machine learning algorithms yet, can you guess how important `Feature 3` is to our data? On one hand, I can imagine that a model might not be great at distinguishing the importance of this feature. Many of the samples have a low value for this feature, and it may be difficult to distinguish one sample from another given this feature. On the other hand, knowing that a sample has a high value for this feature might be informative.

Let's take a look at the same information, using a boxplot instead.

```python
fig, axs = plt.subplots(ncols=7, nrows=2, figsize=(20, 10))
axs = axs.flatten()

for idx, feature in enumerate(X.T):
    axs[idx].boxplot(feature)
    axs[idx].set_title(f'{feature_names[idx]}')

axs[-1].boxplot(y)
_ = axs[-1].set_title(f'Target: MEDV')

plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

```python
# There are better ways of keeping track of this sort of thing 
# with pandas, but since we have been working with numpy up until
# this point, forgive me for indexing in this way. Indexing errors
# are common in programming, so we have to make sure we know what we
# are getting when we slice into an array like this.
proportion_non_retail = X[:, 2]
pupil_teacher_ratio = X[:, 10]

hist_features = {
    'proportion_non_retail': proportion_non_retail, 
    'pupil_teacher_ratio': pupil_teacher_ratio
}

fig, ax = plt.subplots(figsize=(20, 10))

for label, feature in hist_features.items():
    ax.hist(feature, label=label, alpha=0.3)

plt.legend(fontsize='x-large')
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

### How Do These Features Relate to our Target?

Let's take a closer look at some of our features and how they relate to the median home value in Boston. Let's first plot one of features along the x-axis with the median home value along the y-axis.

```python
# Simple approach with state machine approach
plt.figure(figsize=(20, 10))

plt.scatter(
    pupil_teacher_ratio, 
    y, 
    s=100,            # keeping a uniform size for all points
    alpha=0.5,        # transparency for overlapping points
    edgecolor="black" # give definition to each point
)

plt.xlabel("Pupil to Teacher Ratio")
plt.ylabel(r"Median Home Value ($\times \$1000$)")
plt.suptitle("Pupil to Teacher Ratio By Housing Value")
plt.show()
```

Sometimes we want to see how two or more features relate to one another. Let's create a scatter plot showing how the proportion of owner-occupied units built prior to 1940 relates to the weighted distances to five Boston employment centres. We can layer in extra information by coloring our points by our target variable, the median value of homes in Boston. Add layers of design features like color and the size of our plotted points can give us a bit more nuance in our visualizations. Just be careful, we should always consider whether each element in our visualization adds information.

```python
proportion_owner_occupied = X[:, 6]
distances_to_employment = X[:, 7]

# Fancier example using the state machine approach
plt.figure(figsize=(20, 10))

plt.scatter(
    proportion_owner_occupied, 
    distances_to_employment, 
    s=100,        # keeping a uniform size for all points
    c=y,          # color the points by our target
    cmap='magma', # colormaps can make things easier to read
    alpha=0.5     # transparency for overlapping points
)

cbar = plt.colorbar()
cbar.set_label(r"Median Home Value ($\times \$1000$)")

plt.xlabel("Proportion of Owner Occupied Units Built Prior to 1940")
plt.ylabel("Weighted Mean Distance to Employment Center")
plt.suptitle("Owner Occupied Homes and Distance to Employment")
plt.show()
```
